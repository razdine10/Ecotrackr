````python
# ── cellule : définition de la fonction ─────────────────────────────────────
import requests
import pandas as pd

def get_impact(
    activity_id: str,
    amount: float,
    unit: str = "kWh",
    region: str = "FR",
    data_version: str = "2022",
) -> pd.DataFrame:
    """
    Interroge l'endpoint /estimate de ClimaTiq et
    renvoie un DataFrame pandas (une seule ligne).
    """
    endpoint = "https://api.climatiq.io/estimate"
      # ← nom de domaine .io

    payload = {
        "emission_factor": {
            "activity_id":  activity_id,
            "data_version": data_version,
            "region":       region,
        },
        "parameters": {
            "energy":      amount,
            "energy_unit": unit,
        },
    }

    # Requête POST (timeout à 15 s pour éviter un blocage infini)
    resp = requests.post(endpoint, headers=HEADERS, json=payload, timeout=15)
    resp.raise_for_status()                 # stoppe ici si code HTTP ≠ 200

    return pd.json_normalize(resp.json())   # DataFrame 1 ligne
````

````python
import os, requests
from dotenv import load_dotenv
load_dotenv()

API_KEY = os.getenv("IMPACT_CO2_API_KEY")
BASE        = "https://impactco2.fr/api/v1"
THEMA_ID    = 8          # chauffage
SURFACE_M2  = 100        # change ta surface ici

HEADERS = {
    "accept": "application/json",
    **({"Authorization": f"Bearer {API_KEY}"} if API_KEY else {})
}

# 1) Récupération des ECV
resp = requests.get(
    f"{BASE}/thematiques/ecv/{THEMA_ID}",
    params={"detail": 1, "language": "fr"},
    headers=HEADERS, timeout=10
)
resp.raise_for_status()
data = resp.json()

# Debug : affichons la structure des données
print("🔍 Structure des données reçues :")
print(f"Type de data: {type(data)}")
print(f"Contenu brut: {data}")

# Traitement selon la structure réelle
if isinstance(data, dict):
    items = list(data.values())
    print(f"\nData était un dict, converti en liste de {len(items)} éléments")
elif isinstance(data, list):
    items = data
    print(f"\nData était déjà une liste de {len(items)} éléments")
else:
    print(f"\nStructure inattendue : {type(data)}")
    items = []

# Debug : structure du premier élément
if items:
    print(f"\nPremier élément:")
    print(f"Type: {type(items[0])}")
    print(f"Contenu: {items[0]}")
    
    # Si c'est une liste, on essaie de comprendre sa structure
    if isinstance(items[0], list):
        print(f"Longueur de la liste: {len(items[0])}")
        for i, elem in enumerate(items[0][:5]):  # Affiche les 5 premiers éléments
            print(f"  Index {i}: {elem} (type: {type(elem)})")

print(f"\n📊 Émissions pour {SURFACE_M2} m² :\n")

# Traitement adapté selon la structure réelle
for i, o in enumerate(items):
    try:
        if isinstance(o, dict):
            # Structure attendue (dictionnaire)
            name   = o.get("name", f"Élément {i}")
            slug   = o.get("slug", "unknown")
            factor = o.get("ecv", 0)
            co2e   = round(factor * SURFACE_M2, 3)
            print(f"• {name:<45} → {co2e:>8} kg CO₂e "
                  f"(facteur {factor} kg/m², slug: {slug})")
                  
        elif isinstance(o, list):
            # Structure inattendue (liste)
            print(f"• Élément {i} (liste): {o}")
            # Si on peut identifier des patterns dans la liste
            if len(o) >= 3:
                print(f"  → Possible: nom={o[0]}, slug={o[1]}, facteur={o[2]}")
                
        else:
            print(f"• Élément {i} ({type(o)}): {o}")
            
    except Exception as e:
        print(f"• Erreur sur l'élément {i}: {e}")
        print(f"  Contenu: {o}")
````

````python
import json
from typing import Dict, List, Any

def analyze_impact_co2_data(data: Dict[str, Any]) -> None:
    """
    Analyse les données reçues de l'API Impact CO2
    """
    print("🔍 ANALYSE DES DONNÉES IMPACT CO2")
    print("=" * 50)
    
    # Extraire la liste des facteurs d'émission
    if 'data' in data and isinstance(data['data'], list):
        factors = data['data']
        print(f"📊 {len(factors)} facteurs d'émission trouvés\n")
        
        # Analyser chaque facteur
        electricity_factors = []
        all_factors = []
        
        for i, factor in enumerate(factors):
            name = factor.get('name', 'N/A')
            slug = factor.get('slug', 'N/A')
            ecv = factor.get('ecv', 0)  # Émissions de CO2 équivalent
            footprint = factor.get('footprint', 0)  # Empreinte carbone
            
            all_factors.append({
                'index': i,
                'name': name,
                'slug': slug,
                'ecv': ecv,
                'footprint': footprint
            })
            
            # Identifier les facteurs liés à l'électricité
            if 'électrique' in name.lower() or 'electric' in name.lower():
                electricity_factors.append({
                    'name': name,
                    'slug': slug,
                    'ecv': ecv,
                    'footprint': footprint,
                    'index': i
                })
                
        # Afficher tous les facteurs
        print("📋 TOUS LES FACTEURS D'ÉMISSION:")
        for factor in all_factors:
            print(f"   {factor['index']+1:2d}. {factor['name']}")
            print(f"       • Slug: {factor['slug']}")
            print(f"       • ECV: {factor['ecv']} kg CO2e")
            print(f"       • Footprint: {factor['footprint']} kg CO2e")
            print()
        
        # Focus sur l'électricité
        if electricity_factors:
            print("⚡ FACTEURS D'ÉLECTRICITÉ IDENTIFIÉS:")
            for elec in electricity_factors:
                print(f"   📍 {elec['name']}")
                print(f"      • Facteur: {elec['ecv']} kg CO2e/m²")
                print(f"      • Slug: {elec['slug']}")
                print()
                
                # Calculer les équivalences pour l'électricité
                calculate_electricity_equivalents(elec)
        else:
            print("❌ Aucun facteur d'électricité spécifique trouvé")
            
        # Comparaison des différents modes de chauffage
        print("\n🔥 COMPARAISON DES MODES DE CHAUFFAGE (pour 100 m²):")
        sorted_factors = sorted(all_factors, key=lambda x: x['ecv'])
        
        for factor in sorted_factors:
            emissions_100m2 = factor['ecv'] * 100
            print(f"   {factor['name']:<40} {emissions_100m2:>8.0f} kg CO2e")
            
        # Calculer les économies potentielles
        print("\n💰 ÉCONOMIES POTENTIELLES (vs chauffage le plus polluant):")
        worst_case = max(all_factors, key=lambda x: x['ecv'])
        best_case = min(all_factors, key=lambda x: x['ecv'])
        
        for factor in sorted_factors:
            if factor['ecv'] != worst_case['ecv']:
                savings = (worst_case['ecv'] - factor['ecv']) * 100
                percentage = (savings / (worst_case['ecv'] * 100)) * 100
                print(f"   {factor['name']:<40} -{savings:>7.0f} kg CO2e ({percentage:>4.1f}%)")

def calculate_electricity_equivalents(electricity_factor: Dict[str, Any]) -> None:
    """
    Calcule les équivalences pour le facteur d'électricité trouvé
    """
    print(f"🔌 CALCULS D'ÉQUIVALENCES POUR: {electricity_factor['name']}")
    print("─" * 60)
    
    # Le facteur est en kg CO2e/m² pour le chauffage électrique
    # Il faut estimer la consommation électrique par m²
    
    # Estimation basée sur les standards français
    # Chauffage électrique: environ 50-80 kWh/m²/an selon l'isolation
    kwh_per_m2_scenarios = {
        "Logement bien isolé (RT2012)": 50,
        "Logement moyennement isolé": 65,
        "Logement mal isolé": 80
    }
    
    ecv_per_m2 = electricity_factor['ecv']  # kg CO2e/m²
    
    print(f"📊 Facteur de base: {ecv_per_m2} kg CO2e/m² de surface chauffée")
    print()
    
    for scenario, kwh_per_m2 in kwh_per_m2_scenarios.items():
        # Calculer le facteur d'émission par kWh
        if kwh_per_m2 > 0:
            kg_co2_per_kwh = ecv_per_m2 / kwh_per_m2
            
            print(f"🏠 {scenario}:")
            print(f"   • Consommation estimée: {kwh_per_m2} kWh/m²/an")
            print(f"   • Facteur d'émission: {kg_co2_per_kwh:.4f} kg CO2e/kWh")
            print(f"   • Émissions pour 100 kWh: {kg_co2_per_kwh * 100:.2f} kg CO2e")
            print(f"   • Émissions pour 1000 kWh: {kg_co2_per_kwh * 1000:.1f} kg CO2e")
            print()

def convert_heating_to_electricity_factor(heating_data: Dict[str, Any]) -> Dict[str, Any]:
    """
    Convertit les données de chauffage électrique en facteur d'émission électricité pure
    """
    electricity_heating = None
    
    if 'data' in heating_data:
        for item in heating_data['data']:
            if 'électrique' in item.get('name', '').lower():
                electricity_heating = item
                break
    
    if not electricity_heating:
        return None
    
    # Estimation du facteur d'émission de l'électricité
    # Basé sur une consommation moyenne de 65 kWh/m²/an pour le chauffage électrique
    avg_kwh_per_m2 = 65
    heating_emissions_per_m2 = electricity_heating['ecv']  # kg CO2e/m²
    
    electricity_factor = heating_emissions_per_m2 / avg_kwh_per_m2  # kg CO2e/kWh
    
    return {
        'source': 'Impact CO2 - Chauffage électrique',
        'factor': electricity_factor,
        'unit': 'kg CO2e/kWh',
        'region': 'France',
        'method': f'Calculé à partir du chauffage électrique ({heating_emissions_per_m2} kg CO2e/m² ÷ {avg_kwh_per_m2} kWh/m²)',
        'raw_heating_data': electricity_heating
    }

def main_analysis():
    """
    Fonction principale pour analyser vos données
    """
    # Vos données reçues de l'API
    your_data = {
        'data': [
            {'name': 'Chauffage au gaz par m²', 'slug': 'chauffagegaz', 'ecv': 39, 'footprint': 39},
            {'name': 'Chauffage au fioul par m²', 'slug': 'chauffagefioul', 'ecv': 57.17, 'footprint': 57.17},
            {'name': 'Chauffage électrique par m²', 'slug': 'chauffageelectrique', 'ecv': 11.85, 'footprint': 11.85},
            {'name': 'Chauffage avec une pompe à chaleur par m²', 'slug': 'pompeachaleur', 'ecv': 3.95, 'footprint': 3.95},
            {'name': 'Chauffage avec un poêle à granulés par m²', 'slug': 'poeleagranule', 'ecv': 5.64, 'footprint': 5.64},
            {'name': 'Chauffage avec un poêle à bois par m²', 'slug': 'poeleabois', 'ecv': 9.2, 'footprint': 9.2},
            {'name': 'Chauffage via un réseau de chaleur par m²', 'slug': 'reseaudechaleur', 'ecv': 18.67, 'footprint': 18.67}
        ]
    }
    
    # Analyser les données
    analyze_impact_co2_data(your_data)
    
    # Convertir en facteur d'électricité
    print("\n" + "="*70)
    print("🔌 CONVERSION EN FACTEUR D'ÉMISSION ÉLECTRICITÉ")
    print("="*70)
    
    electricity_factor = convert_heating_to_electricity_factor(your_data)
    
    if electricity_factor:
        print(f"✅ Facteur d'émission électricité estimé: {electricity_factor['factor']:.4f} kg CO2e/kWh")
        print(f"📍 Méthode: {electricity_factor['method']}")
        print(f"🌍 Région: {electricity_factor['region']}")
        
        print(f"\n💨 CALCULS FINAUX:")
        print(f"   • 1 kWh: {electricity_factor['factor']:.4f} kg CO2e")
        print(f"   • 100 kWh: {electricity_factor['factor'] * 100:.2f} kg CO2e")
        print(f"   • 1000 kWh: {electricity_factor['factor'] * 1000:.1f} kg CO2e")
        print(f"   • Consommation annuelle moyenne (2500 kWh): {electricity_factor['factor'] * 2500:.1f} kg CO2e")
        
        # Comparaison avec d'autres sources
        print(f"\n📊 COMPARAISON AVEC D'AUTRES SOURCES:")
        print(f"   • Votre calcul Impact CO2: {electricity_factor['factor']:.4f} kg CO2e/kWh")
        print(f"   • Base Carbone ADEME: ~0.0571 kg CO2e/kWh")
        print(f"   • RTE (mix français): ~0.0570 kg CO2e/kWh")
        
        difference = abs(electricity_factor['factor'] - 0.0571)
        print(f"   • Différence: {difference:.4f} kg CO2e/kWh")
        
        if difference < 0.01:
            print("   ✅ Cohérent avec les données officielles")
        else:
            print("   ⚠️ Écart notable - vérifier la méthode de calcul")
    else:
        print("❌ Impossible de calculer le facteur d'électricité")

if __name__ == "__main__":
    main_analysis()
````

````python
import requests, operator, os
from dotenv import load_dotenv

load_dotenv()
HEAD = {
    "accept": "application/json",
    **({"Authorization": f"Bearer "+os.getenv("IMPACT_CO2_API_KEY")} if os.getenv("IMPACT_CO2_API_KEY") else {})
}
BASE = "https://impactco2.fr/api/v1"

def ecv_list(theme_id: int, detail: int = 1):
    r = requests.get(f"{BASE}/thematiques/ecv/{theme_id}",
                     params={"detail": detail, "language": "fr"},
                     headers=HEAD, timeout=10)
    r.raise_for_status()
    data = r.json()["data"]           # <-- déballage
    # si detail=0 ➜ dict ; si dict → liste
    return list(data.values()) if isinstance(data, dict) else data

# 1) Usage numérique (id 10)
num_objects = ecv_list(10)            # list[dict]
streaming   = next(o for o in num_objects if o["slug"] == "streamingvideo")["ecv"]  # kg/h
email       = next(o for o in num_objects if o["slug"] == "email")["ecv"]           # kg/mail

# 2) Chauffage (id 8)
heat_objects = ecv_list(8)
elec_heat    = next(o for o in heat_objects if o["slug"] == "chauffageelectrique")["ecv"]  # kg/m²

# 3) Quantités réelles
usage = [
    ("1 h de streaming",     streaming * 2),
    ("50 e-mails",           email * 100),
    ("100 m² chauffés élec", elec_heat * 10),
]

# 4) Classement
usage.sort(key=operator.itemgetter(1))

print("📊 Empreintes comparées :")
for label, kg in usage:
    print(f"• {label:<25} → {kg:>8.3f} kg CO₂e")
````

````python
# ╔═══════════════════════════════════════════════════════════════════╗
# ║  ANALYSE TRANSPORTS Impact CO₂ (micro-modes + agrégat macro)      ║
# ╚═══════════════════════════════════════════════════════════════════╝
import os, re, requests, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from dotenv import load_dotenv

# ─── 0) Authentification ────────────────────────────────────────────
load_dotenv()
API_KEY = os.getenv("IMPACT_CO2_API_KEY")
assert API_KEY, "⚠️  IMPACT_CO2_API_KEY manquante dans .env"

HEAD = {"accept": "application/json", "Authorization": f"Bearer {API_KEY}"}
BASE = "https://impactco2.fr/api/v1"
THEME_TRANSPORT = 4                     # 4 = thématique « transport »

# ─── 1) Téléchargement des sous-modes (« micro ») ───────────────────
def micro_transport() -> pd.DataFrame:
    r = requests.get(
        f"{BASE}/thematiques/ecv/{THEME_TRANSPORT}",
        params={"detail": 1, "language": "fr"},
        headers=HEAD, timeout=10,
    )
    r.raise_for_status()
    rows = r.json()["data"]
    if isinstance(rows, dict):      # (sécurité si detail=0)
        rows = list(rows.values())
    return (pd.DataFrame(rows)[["name", "slug", "ecv"]]
            .rename(columns={"name": "Objet", "ecv": "kg_CO2e"})
            .assign(kg_CO2e=lambda d: d["kg_CO2e"].astype(float)))

df = micro_transport()

# ─── 2) Mapping slug → macro-mode ───────────────────────────────────
def to_macro(slug: str) -> str:
    mapping = {
        r"avion":                         "Avion",
        r"\b(tgv|ter|rer|intercites|train)\b": "Train",
        r"\b(bus|autocar|gnv)\b":         "Bus/Car",
        r"tram|metro":                    "Tram/Métro",
        r"voiture":                       "Voiture",
        r"moto|scooter":                  "2-roues motorisés",
        r"velo":                          "Vélo",
        r"trottinette":                   "Trottinette",
        r"marche":                        "Marche",
    }
    slug = slug.lower()
    for pat, macro in mapping.items():
        if re.search(pat, slug):
            return macro
    return "Autre"

df["Macro"] = df["slug"].apply(to_macro)

# ─── 3) Agrégation macro pour la seconde figure ─────────────────────
macro_df = (df.groupby("Macro", as_index=False)["kg_CO2e"]
              .mean()
              .sort_values("kg_CO2e", ascending=False))

# ─── 4) Couleurs cohérentes ─────────────────────────────────────────
palette   = sns.color_palette("Set2", len(macro_df))
color_map = dict(zip(macro_df["Macro"], palette))

# ─── 5) Visualisations ──────────────────────────────────────────────
sns.set_style("whitegrid")
plt.rcParams["font.family"] = "DejaVu Sans"

fig, (ax_micro, ax_macro) = plt.subplots(
    1, 2, figsize=(15, 6), gridspec_kw={"width_ratios": [2.2, 1]}
)

# 5-A  Barres horizontales – micro-modes
df_sorted = df.sort_values("kg_CO2e", ascending=False)
sns.barplot(
    data=df_sorted,
    x="kg_CO2e", y="Objet",
    hue="Macro",
    palette=color_map,
    dodge=False, ax=ax_micro,
)
ax_micro.set(
    title="Tous les sous-modes de transport (ordre décroissant)",
    xlabel="kg CO₂e / passager·km (ou équivalent)",
    ylabel="",
)
ax_micro.legend(title="Macro-mode", bbox_to_anchor=(1.02, 1), loc="upper left")

# 5-B  Barres horizontales – macro-modes
sns.barplot(
    data=macro_df,
    x="kg_CO2e", y="Macro",
    hue="Macro",           # nécessaire pour appliquer la palette
    palette=color_map,
    dodge=False, legend=False,
    orient="h", ax=ax_macro,
)
ax_macro.set(
    title="Répartition moyenne de l’empreinte par macro-mode",
    xlabel="kg CO₂e / passager·km (moyenne)",
    ylabel="",
)
for i, v in enumerate(macro_df["kg_CO2e"]):
    ax_macro.text(v + 0.002, i, f"{v:.3f}", va="center")

plt.tight_layout()
plt.show()
````
