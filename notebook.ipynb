````python
# ‚îÄ‚îÄ cellule : d√©finition de la fonction ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
import requests
import pandas as pd

def get_impact(
    activity_id: str,
    amount: float,
    unit: str = "kWh",
    region: str = "FR",
    data_version: str = "2022",
) -> pd.DataFrame:
    """
    Interroge l'endpoint /estimate de ClimaTiq et
    renvoie un DataFrame pandas (une seule ligne).
    """
    endpoint = "https://api.climatiq.io/estimate"
      # ‚Üê nom de domaine .io

    payload = {
        "emission_factor": {
            "activity_id":  activity_id,
            "data_version": data_version,
            "region":       region,
        },
        "parameters": {
            "energy":      amount,
            "energy_unit": unit,
        },
    }

    # Requ√™te POST (timeout √† 15 s pour √©viter un blocage infini)
    resp = requests.post(endpoint, headers=HEADERS, json=payload, timeout=15)
    resp.raise_for_status()                 # stoppe ici si code HTTP ‚â† 200

    return pd.json_normalize(resp.json())   # DataFrame 1 ligne
````

````python
# --- Cellule 2 : d√©finition de la fonction -----------------------------------
def search_factors(query: str, *, limit: int = 20) -> pd.DataFrame:
    """
    Recherche des 'emission_factors' dans l'API Climatiq
    et renvoie un DataFrame des r√©sultats.
    """
    url     = "https://api.climatiq.io/search"
    params  = {"query": query, "limit": limit}
    r       = requests.get(url, headers=HEADERS, params=params, timeout=15)
    r.raise_for_status()               # l√®ve une erreur si code HTTP ‚â† 200
    return pd.json_normalize(r.json()["results"])
````

````python
# ‚îÄ‚îÄ Cellule 1 : fonction de recherche ---------------------------------
import requests, pandas as pd, os
from dotenv import load_dotenv

load_dotenv()
HEADERS = {"Authorization": f"Bearer {os.getenv('CLIMATIQ_API_KEY')}"}

def search_factors(query: str,
                   page:int = 1,
                   per_page:int = 50) -> pd.DataFrame:
    """
    Renvoie un DataFrame d'√©mission-factors pour le mot-cl√© `query`.
    - page        : num√©ro de page (>=1)
    - per_page    : 1-100 r√©sultats par page
    """
    url    = "https://api.climatiq.io/search"
    params = {
        "query"            : query,
        "page"             : page,
        "results_per_page" : per_page      # <- la bonne clef !
    }
    r = requests.get(url, headers=HEADERS, params=params, timeout=15)
    r.raise_for_status()                  # stoppe si code HTTP ‚â† 200
    return pd.json_normalize(r.json()["results"])
````

````python
def search_factors(
    query: str,
    page: int = 1,
    results_per_page: int = 20
) -> pd.DataFrame:
    """
    Interroge l'endpoint /search de Climatiq.
    Renvoie un DataFrame des 'results'.
    """
    url = "https://api.climatiq.io/search"
    params = {
        "query": query,
        "page": page,
        "results_per_page": results_per_page
    }
    r = requests.get(url, headers=HEADERS, params=params, timeout=15)
    r.raise_for_status()
    return pd.json_normalize(r.json()["results"])
````

````python
import json
from typing import Dict, List, Any

def analyze_impact_co2_data(data: Dict[str, Any]) -> None:
    """
    Analyse les donn√©es re√ßues de l'API Impact CO2
    """
    print("üîç ANALYSE DES DONN√âES IMPACT CO2")
    print("=" * 50)
    
    # Extraire la liste des facteurs d'√©mission
    if 'data' in data and isinstance(data['data'], list):
        factors = data['data']
        print(f"üìä {len(factors)} facteurs d'√©mission trouv√©s\n")
        
        # Analyser chaque facteur
        electricity_factors = []
        all_factors = []
        
        for i, factor in enumerate(factors):
            name = factor.get('name', 'N/A')
            slug = factor.get('slug', 'N/A')
            ecv = factor.get('ecv', 0)  # √âmissions de CO2 √©quivalent
            footprint = factor.get('footprint', 0)  # Empreinte carbone
            
            all_factors.append({
                'index': i,
                'name': name,
                'slug': slug,
                'ecv': ecv,
                'footprint': footprint
            })
            
            # Identifier les facteurs li√©s √† l'√©lectricit√©
            if '√©lectrique' in name.lower() or 'electric' in name.lower():
                electricity_factors.append({
                    'name': name,
                    'slug': slug,
                    'ecv': ecv,
                    'footprint': footprint,
                    'index': i
                })
                
        # Afficher tous les facteurs
        print("üìã TOUS LES FACTEURS D'√âMISSION:")
        for factor in all_factors:
            print(f"   {factor['index']+1:2d}. {factor['name']}")
            print(f"       ‚Ä¢ Slug: {factor['slug']}")
            print(f"       ‚Ä¢ ECV: {factor['ecv']} kg CO2e")
            print(f"       ‚Ä¢ Footprint: {factor['footprint']} kg CO2e")
            print()
        
        # Focus sur l'√©lectricit√©
        if electricity_factors:
            print("‚ö° FACTEURS D'√âLECTRICIT√â IDENTIFI√âS:")
            for elec in electricity_factors:
                print(f"   üìç {elec['name']}")
                print(f"      ‚Ä¢ Facteur: {elec['ecv']} kg CO2e/m¬≤")
                print(f"      ‚Ä¢ Slug: {elec['slug']}")
                print()
                
                # Calculer les √©quivalences pour l'√©lectricit√©
                calculate_electricity_equivalents(elec)
        else:
            print("‚ùå Aucun facteur d'√©lectricit√© sp√©cifique trouv√©")
            
        # Comparaison des diff√©rents modes de chauffage
        print("\nüî• COMPARAISON DES MODES DE CHAUFFAGE (pour 100 m¬≤):")
        sorted_factors = sorted(all_factors, key=lambda x: x['ecv'])
        
        for factor in sorted_factors:
            emissions_100m2 = factor['ecv'] * 100
            print(f"   {factor['name']:<40} {emissions_100m2:>8.0f} kg CO2e")
            
        # Calculer les √©conomies potentielles
        print("\nüí∞ √âCONOMIES POTENTIELLES (vs chauffage le plus polluant):")
        worst_case = max(all_factors, key=lambda x: x['ecv'])
        best_case = min(all_factors, key=lambda x: x['ecv'])
        
        for factor in sorted_factors:
            if factor['ecv'] != worst_case['ecv']:
                savings = (worst_case['ecv'] - factor['ecv']) * 100
                percentage = (savings / (worst_case['ecv'] * 100)) * 100
                print(f"   {factor['name']:<40} -{savings:>7.0f} kg CO2e ({percentage:>4.1f}%)")

def calculate_electricity_equivalents(electricity_factor: Dict[str, Any]) -> None:
    """
    Calcule les √©quivalences pour le facteur d'√©lectricit√© trouv√©
    """
    print(f"üîå CALCULS D'√âQUIVALENCES POUR: {electricity_factor['name']}")
    print("‚îÄ" * 60)
    
    # Le facteur est en kg CO2e/m¬≤ pour le chauffage √©lectrique
    # Il faut estimer la consommation √©lectrique par m¬≤
    
    # Estimation bas√©e sur les standards fran√ßais
    # Chauffage √©lectrique: environ 50-80 kWh/m¬≤/an selon l'isolation
    kwh_per_m2_scenarios = {
        "Logement bien isol√© (RT2012)": 50,
        "Logement moyennement isol√©": 65,
        "Logement mal isol√©": 80
    }
    
    ecv_per_m2 = electricity_factor['ecv']  # kg CO2e/m¬≤
    
    print(f"üìä Facteur de base: {ecv_per_m2} kg CO2e/m¬≤ de surface chauff√©e")
    print()
    
    for scenario, kwh_per_m2 in kwh_per_m2_scenarios.items():
        # Calculer le facteur d'√©mission par kWh
        if kwh_per_m2 > 0:
            kg_co2_per_kwh = ecv_per_m2 / kwh_per_m2
            
            print(f"üè† {scenario}:")
            print(f"   ‚Ä¢ Consommation estim√©e: {kwh_per_m2} kWh/m¬≤/an")
            print(f"   ‚Ä¢ Facteur d'√©mission: {kg_co2_per_kwh:.4f} kg CO2e/kWh")
            print(f"   ‚Ä¢ √âmissions pour 100 kWh: {kg_co2_per_kwh * 100:.2f} kg CO2e")
            print(f"   ‚Ä¢ √âmissions pour 1000 kWh: {kg_co2_per_kwh * 1000:.1f} kg CO2e")
            print()

def convert_heating_to_electricity_factor(heating_data: Dict[str, Any]) -> Dict[str, Any]:
    """
    Convertit les donn√©es de chauffage √©lectrique en facteur d'√©mission √©lectricit√© pure
    """
    electricity_heating = None
    
    if 'data' in heating_data:
        for item in heating_data['data']:
            if '√©lectrique' in item.get('name', '').lower():
                electricity_heating = item
                break
    
    if not electricity_heating:
        return None
    
    # Estimation du facteur d'√©mission de l'√©lectricit√©
    # Bas√© sur une consommation moyenne de 65 kWh/m¬≤/an pour le chauffage √©lectrique
    avg_kwh_per_m2 = 65
    heating_emissions_per_m2 = electricity_heating['ecv']  # kg CO2e/m¬≤
    
    electricity_factor = heating_emissions_per_m2 / avg_kwh_per_m2  # kg CO2e/kWh
    
    return {
        'source': 'Impact CO2 - Chauffage √©lectrique',
        'factor': electricity_factor,
        'unit': 'kg CO2e/kWh',
        'region': 'France',
        'method': f'Calcul√© √† partir du chauffage √©lectrique ({heating_emissions_per_m2} kg CO2e/m¬≤ √∑ {avg_kwh_per_m2} kWh/m¬≤)',
        'raw_heating_data': electricity_heating
    }

def main_analysis():
    """
    Fonction principale pour analyser vos donn√©es
    """
    # Vos donn√©es re√ßues de l'API
    your_data = {
        'data': [
            {'name': 'Chauffage au gaz par m¬≤', 'slug': 'chauffagegaz', 'ecv': 39, 'footprint': 39},
            {'name': 'Chauffage au fioul par m¬≤', 'slug': 'chauffagefioul', 'ecv': 57.17, 'footprint': 57.17},
            {'name': 'Chauffage √©lectrique par m¬≤', 'slug': 'chauffageelectrique', 'ecv': 11.85, 'footprint': 11.85},
            {'name': 'Chauffage avec une pompe √† chaleur par m¬≤', 'slug': 'pompeachaleur', 'ecv': 3.95, 'footprint': 3.95},
            {'name': 'Chauffage avec un po√™le √† granul√©s par m¬≤', 'slug': 'poeleagranule', 'ecv': 5.64, 'footprint': 5.64},
            {'name': 'Chauffage avec un po√™le √† bois par m¬≤', 'slug': 'poeleabois', 'ecv': 9.2, 'footprint': 9.2},
            {'name': 'Chauffage via un r√©seau de chaleur par m¬≤', 'slug': 'reseaudechaleur', 'ecv': 18.67, 'footprint': 18.67}
        ]
    }
    
    # Analyser les donn√©es
    analyze_impact_co2_data(your_data)
    
    # Convertir en facteur d'√©lectricit√©
    print("\n" + "="*70)
    print("üîå CONVERSION EN FACTEUR D'√âMISSION √âLECTRICIT√â")
    print("="*70)
    
    electricity_factor = convert_heating_to_electricity_factor(your_data)
    
    if electricity_factor:
        print(f"‚úÖ Facteur d'√©mission √©lectricit√© estim√©: {electricity_factor['factor']:.4f} kg CO2e/kWh")
        print(f"üìç M√©thode: {electricity_factor['method']}")
        print(f"üåç R√©gion: {electricity_factor['region']}")
        
        print(f"\nüí® CALCULS FINAUX:")
        print(f"   ‚Ä¢ 1 kWh: {electricity_factor['factor']:.4f} kg CO2e")
        print(f"   ‚Ä¢ 100 kWh: {electricity_factor['factor'] * 100:.2f} kg CO2e")
        print(f"   ‚Ä¢ 1000 kWh: {electricity_factor['factor'] * 1000:.1f} kg CO2e")
        print(f"   ‚Ä¢ Consommation annuelle moyenne (2500 kWh): {electricity_factor['factor'] * 2500:.1f} kg CO2e")
        
        # Comparaison avec d'autres sources
        print(f"\nüìä COMPARAISON AVEC D'AUTRES SOURCES:")
        print(f"   ‚Ä¢ Votre calcul Impact CO2: {electricity_factor['factor']:.4f} kg CO2e/kWh")
        print(f"   ‚Ä¢ Base Carbone ADEME: ~0.0571 kg CO2e/kWh")
        print(f"   ‚Ä¢ RTE (mix fran√ßais): ~0.0570 kg CO2e/kWh")
        
        difference = abs(electricity_factor['factor'] - 0.0571)
        print(f"   ‚Ä¢ Diff√©rence: {difference:.4f} kg CO2e/kWh")
        
        if difference < 0.01:
            print("   ‚úÖ Coh√©rent avec les donn√©es officielles")
        else:
            print("   ‚ö†Ô∏è √âcart notable - v√©rifier la m√©thode de calcul")
    else:
        print("‚ùå Impossible de calculer le facteur d'√©lectricit√©")

if __name__ == "__main__":
    main_analysis()
````

````python
# ‚ïî‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïó
# ‚ïë Empreinte CO‚ÇÇe ‚Äì analyse MICRO (objets d'une th√©matique)          ‚ïë
# ‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù
import os, requests, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from dotenv import load_dotenv

# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ 0) Auth & config ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
load_dotenv()
API_KEY = os.getenv("IMPACT_CO2_API_KEY")
assert API_KEY, "‚ö†Ô∏è  IMPACT_CO2_API_KEY manquante !"

HEAD = {"accept": "application/json", "Authorization": f"Bearer {API_KEY}"}
BASE = "https://impactco2.fr/api/v1"
THEMES = {1: "Num√©rique", 6: "√âlectrom√©nager", 8: "Chauffage"}
TOP_N  = 10     # nb de barres par graphique

def micro_items(theme_id: int) -> list[dict]:
    """Liste d'objets (detail=1) pour un th√®me."""
    r = requests.get(
        f"{BASE}/thematiques/ecv/{theme_id}",
        params={"detail": 1, "language": "fr"},
        headers=HEAD, timeout=10,
    )
    r.raise_for_status()
    data = r.json()["data"]
    return list(data.values()) if isinstance(data, dict) else data

# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ 1) R√©cup√©ration de TOUTES les lignes micro ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
rows = []
for tid, tname in THEMES.items():
    for item in micro_items(tid):
        rows.append(
            {
                "Th√©matique": tname,
                "Objet":      item["name"],
                "kg_CO2e":    float(item["ecv"]),
            }
        )

df = pd.DataFrame(rows).sort_values("kg_CO2e", ascending=False).reset_index(drop=True)

# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ 2) Aper√ßu tabulaire ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
display(df)

# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ 3-A) Top 15 global (tous th√®mes) ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
plt.figure(figsize=(8, 4))
sns.barplot(
    data=df.nlargest(15, "kg_CO2e"),
    x="kg_CO2e", y="Objet", hue="Th√©matique",
    palette="Set2", dodge=False
)
plt.xlabel("kg CO‚ÇÇe")
plt.ylabel("")
plt.title("Top 15 objets les plus √©metteurs (3 th√©matiques confondues)")
plt.tight_layout()
plt.show()

# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ 3-B) Top N par th√©matique ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
fig, axes = plt.subplots(1, len(THEMES), figsize=(14, 4), sharex=False)
for ax, (tname, sub) in zip(axes, df.groupby("Th√©matique")):
    sns.barplot(
        data=sub.nlargest(TOP_N, "kg_CO2e"),
        x="kg_CO2e", y="Objet",
        color="#4C72B0", ax=ax
    )
    ax.set_title(tname)
    ax.set_xlabel("kg CO‚ÇÇe")
    ax.set_ylabel("")
    for i, v in enumerate(sub.nlargest(TOP_N, "kg_CO2e")["kg_CO2e"]):
        ax.text(v + 0.5, i, f"{v:.1f}", va="center")
plt.suptitle("R√©partition micro : objets les plus √©metteurs par th√©matique", y=1.02)
plt.tight_layout()
plt.show()
````

````python
# üìä Simulation d'Utilisation R√©elle de l'Application

print("üéØ SIMULATION D'UTILISATION ECOTRACKR")
print("=" * 50)

# Simulation d'un utilisateur type sur une semaine
print("üë§ Profil utilisateur : D√©veloppeur parisien, 28 ans")
print("üìÖ P√©riode simul√©e : 7 jours d'activit√©s")

import random
from datetime import datetime, timedelta

# Donn√©es de simulation bas√©es sur les facteurs valid√©s
activities_simulation = []
start_date = datetime.now() - timedelta(days=7)

# Transport (5 jours de travail)
for day in range(5):
    date = start_date + timedelta(days=day)
    # Aller-retour domicile-travail en m√©tro
    activities_simulation.append({
        "date": date.strftime("%d/%m"),
        "category": "üöó Transport", 
        "activity": "M√©tro/RER",
        "quantity": 24,  # km aller-retour
        "unit": "km",
        "factor": 0.003,
        "emissions": 24 * 0.003
    })

# Week-end : sortie en voiture
weekend_trip = {
    "date": (start_date + timedelta(days=5)).strftime("%d/%m"),
    "category": "üöó Transport",
    "activity": "Voiture essence", 
    "quantity": 120,  # sortie en province
    "unit": "km",
    "factor": 0.195,
    "emissions": 120 * 0.195
}
activities_simulation.append(weekend_trip)

# √ânergie domestique quotidienne
for day in range(7):
    date = (start_date + timedelta(days=day)).strftime("%d/%m")
    # Consommation √©lectrique quotidienne
    activities_simulation.append({
        "date": date,
        "category": "‚ö° √ânergie",
        "activity": "√âlectricit√© (mix FR)",
        "quantity": 12,  # kWh/jour appartement
        "unit": "kWh", 
        "factor": 0.057,
        "emissions": 12 * 0.057
    })

# Alimentation (quelques repas)
food_items = [
    {"name": "B≈ìuf", "qty": 0.3, "factor": 25.0},
    {"name": "Volaille", "qty": 0.5, "factor": 4.57},
    {"name": "L√©gumes", "qty": 1.2, "factor": 0.44},
    {"name": "Fromage", "qty": 0.1, "factor": 9.79}
]

for i, food in enumerate(food_items):
    date = (start_date + timedelta(days=i)).strftime("%d/%m")
    activities_simulation.append({
        "date": date,
        "category": "üçΩÔ∏è Alimentation",
        "activity": food["name"],
        "quantity": food["qty"],
        "unit": "kg",
        "factor": food["factor"], 
        "emissions": food["qty"] * food["factor"]
    })

# Num√©rique quotidien
for day in range(7):
    date = (start_date + timedelta(days=day)).strftime("%d/%m")
    # Usage ordinateur (8h de travail)
    activities_simulation.append({
        "date": date,
        "category": "üíª Num√©rique",
        "activity": "Ordinateur portable",
        "quantity": 8,
        "unit": "h",
        "factor": 0.074,
        "emissions": 8 * 0.074
    })

# Cr√©ation du DataFrame
df_simulation = pd.DataFrame(activities_simulation)

# Affichage d√©taill√©
print(f"\nüìã {len(df_simulation)} activit√©s enregistr√©es")
print("\nüîç D√âTAIL DES ACTIVIT√âS:")

for _, activity in df_simulation.iterrows():
    print(f"  üìÖ {activity['date']} | {activity['category']} | "
          f"{activity['activity']:<20} | {activity['quantity']:>6.1f} {activity['unit']:<4} | "
          f"{activity['emissions']:>6.3f} kg CO‚ÇÇe")

# Calculs d'agr√©gation 
total_emissions = df_simulation['emissions'].sum()
daily_average = total_emissions / 7
weekly_by_category = df_simulation.groupby('category')['emissions'].sum().sort_values(ascending=False)

print(f"\nüìä BILAN HEBDOMADAIRE:")
print(f"  ‚Ä¢ Total √©missions : {total_emissions:.2f} kg CO‚ÇÇe")
print(f"  ‚Ä¢ Moyenne quotidienne : {daily_average:.2f} kg CO‚ÇÇe/jour")
print(f"  ‚Ä¢ Projection annuelle : {daily_average * 365:.0f} kg CO‚ÇÇe/an")

print(f"\nüèÜ R√âPARTITION PAR CAT√âGORIE:")
for category, emissions in weekly_by_category.items():
    percentage = (emissions / total_emissions) * 100
    print(f"  ‚Ä¢ {category:<20} : {emissions:>6.2f} kg CO‚ÇÇe ({percentage:>5.1f}%)")

# Graphique de synth√®se
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Graphique 1: √âmissions quotidiennes
daily_emissions = df_simulation.groupby('date')['emissions'].sum()
ax1.bar(daily_emissions.index, daily_emissions.values, alpha=0.8, color='steelblue')
ax1.set_title('üéØ √âmissions Quotidiennes Simul√©es', fontweight='bold')
ax1.set_xlabel('Date')
ax1.set_ylabel('kg CO‚ÇÇe')
ax1.tick_params(axis='x', rotation=45)
ax1.grid(axis='y', alpha=0.3)

# Ajout des valeurs sur les barres
for i, v in enumerate(daily_emissions.values):
    ax1.text(i, v + 0.1, f'{v:.1f}', ha='center', va='bottom')

# Graphique 2: R√©partition par cat√©gorie  
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']
wedges, texts, autotexts = ax2.pie(weekly_by_category.values, 
                                   labels=[cat.split()[1] for cat in weekly_by_category.index],
                                   autopct='%1.1f%%', 
                                   colors=colors, 
                                   startangle=90)
ax2.set_title('üéØ R√©partition par Cat√©gorie', fontweight='bold')

plt.tight_layout()
plt.show()

print(f"\nüí° INSIGHTS AUTOMATIQUES ECOTRACKR:")
print(f"  ‚Ä¢ üöó Le transport repr√©sente {(weekly_by_category['üöó Transport']/total_emissions)*100:.0f}% des √©missions")
print(f"  ‚Ä¢ ‚ö° Optimisation possible: chauffage plus efficace")
print(f"  ‚Ä¢ üçΩÔ∏è Alimentation: impact mod√©r√©, bonnes habitudes") 
print(f"  ‚Ä¢ üíª Num√©rique: usage raisonnable, rien √† signaler")

print(f"\nüéØ L'application EcoTrackr calcule ces m√©triques en temps r√©el")
print(f"   et propose des actions personnalis√©es pour r√©duire l'empreinte")
````

````python
# ‚ïî‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïó
# ‚ïë  ANALYSE TRANSPORTS Impact CO‚ÇÇ (micro-modes + agr√©gat macro)      ‚ïë
# ‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù
import os, re, requests, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from dotenv import load_dotenv

# ‚îÄ‚îÄ‚îÄ 0) Authentification ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
load_dotenv()
API_KEY = os.getenv("IMPACT_CO2_API_KEY")
assert API_KEY, "‚ö†Ô∏è  IMPACT_CO2_API_KEY manquante dans .env"

HEAD = {"accept": "application/json", "Authorization": f"Bearer {API_KEY}"}
BASE = "https://impactco2.fr/api/v1"
THEME_TRANSPORT = 4                     # 4 = th√©matique ¬´ transport ¬ª

# ‚îÄ‚îÄ‚îÄ 1) T√©l√©chargement des sous-modes (¬´ micro ¬ª) ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
def micro_transport() -> pd.DataFrame:
    r = requests.get(
        f"{BASE}/thematiques/ecv/{THEME_TRANSPORT}",
        params={"detail": 1, "language": "fr"},
        headers=HEAD, timeout=10,
    )
    r.raise_for_status()
    rows = r.json()["data"]
    if isinstance(rows, dict):      # (s√©curit√© si detail=0)
        rows = list(rows.values())
    return (pd.DataFrame(rows)[["name", "slug", "ecv"]]
            .rename(columns={"name": "Objet", "ecv": "kg_CO2e"})
            .assign(kg_CO2e=lambda d: d["kg_CO2e"].astype(float)))

df = micro_transport()

# ‚îÄ‚îÄ‚îÄ 2) Mapping slug ‚Üí macro-mode ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
def to_macro(slug: str) -> str:
    mapping = {
        r"avion":                         "Avion",
        r"\b(tgv|ter|rer|intercites|train)\b": "Train",
        r"\b(bus|autocar|gnv)\b":         "Bus/Car",
        r"tram|metro":                    "Tram/M√©tro",
        r"voiture":                       "Voiture",
        r"moto|scooter":                  "2-roues motoris√©s",
        r"velo":                          "V√©lo",
        r"trottinette":                   "Trottinette",
        r"marche":                        "Marche",
    }
    slug = slug.lower()
    for pat, macro in mapping.items():
        if re.search(pat, slug):
            return macro
    return "Autre"

df["Macro"] = df["slug"].apply(to_macro)

# ‚îÄ‚îÄ‚îÄ 3) Agr√©gation macro pour la seconde figure ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
macro_df = (df.groupby("Macro", as_index=False)["kg_CO2e"]
              .mean()
              .sort_values("kg_CO2e", ascending=False))

# ‚îÄ‚îÄ‚îÄ 4) Couleurs coh√©rentes ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
palette   = sns.color_palette("Set2", len(macro_df))
color_map = dict(zip(macro_df["Macro"], palette))

# ‚îÄ‚îÄ‚îÄ 5) Visualisations ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
sns.set_style("whitegrid")
plt.rcParams["font.family"] = "DejaVu Sans"

fig, (ax_micro, ax_macro) = plt.subplots(
    1, 2, figsize=(15, 6), gridspec_kw={"width_ratios": [2.2, 1]}
)

# 5-A  Barres horizontales ‚Äì micro-modes
df_sorted = df.sort_values("kg_CO2e", ascending=False)
sns.barplot(
    data=df_sorted,
    x="kg_CO2e", y="Objet",
    hue="Macro",
    palette=color_map,
    dodge=False, ax=ax_micro,
)
ax_micro.set(
    title="Tous les sous-modes de transport (ordre d√©croissant)",
    xlabel="kg CO‚ÇÇe / passager¬∑km (ou √©quivalent)",
    ylabel="",
)
ax_micro.legend(title="Macro-mode", bbox_to_anchor=(1.02, 1), loc="upper left")

# 5-B  Barres horizontales ‚Äì macro-modes
sns.barplot(
    data=macro_df,
    x="kg_CO2e", y="Macro",
    hue="Macro",           # n√©cessaire pour appliquer la palette
    palette=color_map,
    dodge=False, legend=False,
    orient="h", ax=ax_macro,
)
ax_macro.set(
    title="R√©partition moyenne de l‚Äôempreinte par macro-mode",
    xlabel="kg CO‚ÇÇe / passager¬∑km (moyenne)",
    ylabel="",
)
for i, v in enumerate(macro_df["kg_CO2e"]):
    ax_macro.text(v + 0.002, i, f"{v:.3f}", va="center")

plt.tight_layout()
plt.show()
````
